```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# import warnings

import warnings 
warnings.filterwarnings('ignore')


# Load the dataset

data = pd.read_csv('Customers.csv')

# Convert SignupDate to datetime
data['SignupDate'] = pd.to_datetime(data['SignupDate'])

# Basic Statistics and Structure
print("Dataset Information:")
data.info()
print("\nSummary Statistics:")
print(data.describe(include='all'))
print("\nMissing Values:")
print(data.isnull().sum())

# Unique values in categorical columns
print("\nUnique Regions:", data['Region'].unique())
print("Unique Signup Dates:", data['SignupDate'].nunique())

# Distribution of Regions
plt.figure(figsize=(8, 5))
sns.countplot(x='Region', data=data, palette='viridis')
plt.title('Customer Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()

# Signup Trends Over Time
plt.figure(figsize=(12, 6))
data.groupby('SignupDate').size().plot(kind='line', marker='o', color='b')
plt.title('Signups Over Time')
plt.xlabel('Signup Date')
plt.ylabel('Number of Signups')
plt.grid(True)
plt.show()

# Summary of Region Counts
region_counts = data['Region'].value_counts()
print("\nCustomer Count by Region:")
print(region_counts)
```
